```python
# importing modules
import pandas as pd
import re
from requests import get
from bs4 import BeautifulSoup

# getting url and converting into bs4 element
url = "https://automattic.com/map/"
response = get(url)
html_soup = BeautifulSoup(response.text, 'html.parser')


# manually needed to extract the javascript 'var' and save it as text file
# reading the file
with open('tt.txt') as f:
    data = f.readlines()
    
    
# finding all members name, role, latitude and longitude with some text preprocessing. 
names = []
role = []
latitude = []
longitude = []

for n in data:
    if '"name":' in n:
        n = n.replace(r'"name": ',"")
        n = n.replace(',\n', "")
        n = n.replace("'", "")
        names.append(n[12:])

    if '"role":' in n:
        n = n.replace(r'"role": ',"")
        n = n.replace(',\n', "")
        n = n.replace("'", "")
        role.append(n[12:])
        
    if '"latitude":' in n:
        n = n.replace(r'"latitude": ',"")
        n = n.replace(',\n', "")
        n = n.replace("'", "")
        latitude.append(n[12:])
        
    if '"longitude": ' in n:
        n = n.replace(r'"longitude": ',"")
        n = n.replace(',\n', "")
        n = n.replace('\n',"")
        n = n.replace("'", "")
        longitude.append(n[12:])



automattic_data = pd.DataFrame({'names': names,
                               'Title': role,
                               'latitude': latitude,
                               'longitude': longitude,
                              })
automattic_data.to_csv('automattic.csv')

# checking data
automattic_data.head()
```
